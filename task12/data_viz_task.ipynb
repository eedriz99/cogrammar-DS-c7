```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', keep_default_na=False)
car_data.sample(5)
```

```python
car_data.columns
```

```python
car_data.info()
```

### 1. Boxplot

```python
# Boxplot for revs per mile
plt.figure(figsize=(20, 20))
sns.boxplot(car_data, x='RPM', y='Manufacturer')
plt.ylabel('Manufacturer')
plt.xlabel('Revs per mile (RPM)')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Mazda have the highest revs per mile as its whisker to the right is the longest at about 6500rpm

### 2. Histogram

```python
# Histogram of MPG
car_data_MPG = pd.DataFrame({'highway': car_data['MPG.highway'], 'city': car_data['MPG.city']})
sns.histplot(car_data_MPG)
plt.xlabel('Miles per gallon (MPG)')
plt.ylabel('Frequency')
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
Generally, drives in highway are more fuel efficient, the highest bar for the highway mile per gallon is to the right of that of the city. which shows most vehicles on the highway cover more miles per gallon than most in the city.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure(figsize=(20, 10))
sns.lineplot(car_data, x='Wheelbase', y='Turn.circle')
plt.xlabel('Wheel base')
plt.ylabel('Turning circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
The lineplot shows a positive trend, which shows that the larger the wheelbase, the higher the turning circle gets. At around 110 wheelbase, it can be noticed that the turning circle remains constant.

### 4. Barplot

```python
# Barplot for mean horsepower

sns.barplot(car_data, x='Type', y='Horsepower')
plt.xlabel('Car type')
plt.ylabel('Average horse power')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
from the barplot, it can be deducted that larger cars tend to have more horsepower on average
