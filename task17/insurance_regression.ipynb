## Import packages and libraries

```python
# Data manipulation libraries

import pandas as pd
import numpy as np

# Data visualization libraries

import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msn

# ML libraries

```

```python
df = pd.read_csv('./insurance.csv')
```

```python
df.sample(5)
```

```python
# Check for missing values
msn.matrix(df)
```

```python
df.describe()
```

No missing values.

```python
corr = df.select_dtypes(include=['int64', 'float64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

from here, we can see that the strongest correlation amongst the numerical (independent) variables and the target variable is `.3` which is the variable age.

```python
x = df['age'].to_numpy().reshape(-1, 1)
y = df['charges'].to_numpy().reshape(-1, 1)

print(f'x = {x.shape} \n y = {y.shape}')
# print(f'y = charges\n', y.shape)
```

```python
sns.scatterplot(df, x='age', y='charges')
plt.show()
```

```python
# Import ML libraries

from sklearn.linear_model import LinearRegression
# from sklearn.model_selection import train_test_split
```

```python
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
```

```python
plt.figure()
sns.scatterplot(df, x='age', y='charges')
plt.plot(x, y_pred)
# plt.plot(x_test, y_test)
```
