```python
from sklearn.datasets import load_diabetes
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = load_diabetes(as_frame=True).data[['s1', 's2']]
df
```

```python
plt.figure()
sns.scatterplot(data=df, x='s1', y='s2', palette='hsl', marker='x')
plt.show()
plt.close()
```

```python
df.corr()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
linear_reg = LinearRegression()
```

```python
x = df.drop('s2', axis=1).to_numpy()
# x = df['s1'].values.reshape((-1, 1))
# x.shape

y = df['s2'].to_numpy().reshape((-1, 1))
y.shape
```

```python
tts = train_test_split(x, y, test_size=0.15)
x_train, x_test, y_train, y_test = tts
```

```python
linear_reg.fit(x_train, y_train)
y_pred = linear_reg.predict(x_test)
# y_pred
```

```python
plt.figure()
sns.scatterplot(data=df, x='s1', y='s2')
plt.plot(x_test, y_pred)
# plt.plot(y_test, color='red')
plt.show()
plt.close()
```
